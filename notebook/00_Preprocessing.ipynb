# Goal

This notebook is intended to generate two sets of training/test data for each classifier (Control vs. sPTD and Control vs. PPROM). 

The "raw" data I will be using is from `eset_SC2_v20.RData` file. I have already opened it up in RStudio and saved (`write.csv`) the two objects. The first object `esetSC2` has been saved as `preterm_data.csv`, and the second object `anoSC2` has been saved as `preterm_annotation.csv`.

```python
import pandas as pd
```

```python
data = pd.read_csv("../data/raw/preterm_data.csv", index_col = 0)
print("There are {} genes and {} samples in preterm_data.csv.".format(data.shape[0], data.shape[1]))
data.head()
```

```python
annotation = pd.read_csv("../data/raw/preterm_annotation.csv", index_col = 0)
print("There are {} samples and {} columns in preterm_annotation.csv.".format(annotation.shape[0], annotation.shape[1]))
annotation.tail(15)
```

## Notes and Observations

1. SampleID is unique, but same IndividualID can appear multiple times. This is because some women have multiple samples at different GA (gestational age). 
2. GADel is the gestational age delivery, but this information is not available for test data. 
3. TTD is simply the weeks from sample to delivery, or GADel - GA. 

I think we need to make predictions for individuals, not by samples. Since the gestational age is given, perhaps we may be able to ensemble predictions by a weighted average depending on the gestational age. 

# Separating out the Test Data

First I will separate out the test samples from `data` and `annotation` dataframes.

```python
train_and_test_annotation_indices = annotation.groupby("Train").groups
train_and_test_annotation_indices
```

```python
test_annotation_indices = train_and_test_annotation[0]
test_annotation = annotation.iloc[test_annotation_indices - 1] # Minus 1 is for 0-based counting in Python
assert test_annotation["Train"].sum() == 0
test_annotation.head()
```

```python
train_annotation_indices = train_and_test_annotation_indices[1]
train_annotation = annotation.iloc[train_annotation_indices - 1] # Minus 1 is for 0-based counting in Python
assert train_annotation["Train"].sum() == 435
train_annotation.head()
```

Now let's subset the `data` Dataframe. I am going to tranpose the dataframe because I prefer to work with `samples-by-genes` matrix.

```python
data = data.transpose()
print(data.shape)
data.head()
```

```python
test_SampleIDs = test_annotation["SampleID"]
test_data = data.loc[list(test_SampleIDs)]
print("test_data has {} samples and {} genes.".format(test_data.shape[0], test_data.shape[1]))
```

```python
train_SampleIDs = train_annotation["SampleID"]
train_data = data.loc[list(train_SampleIDs)]
print("train_data has {} samples and {} genes.".format(train_data.shape[0], train_data.shape[1]))
```

## Saving Test Data 

Let's save the test data as `X_test.csv` in the `data/processed` folder. I will also save a `torch.Tensor` version in `data/torch` folder to easily access it for inference later.

I will further separate the `train_data` for each of the two classifiers before saving. 

```python
test_data.to_csv("../data/processed/X_test.csv")
```

```python
import numpy as np
import torch
```

```python
X_test_np = np.float32(test_data.values)
X_test = torch.from_numpy(X_test_np)
torch.save(X_test, "../data/torch/X_test.pt")
```

# Side observation: Class imbalance

```python
train_annotation["Group"].value_counts()
```

It was expected but I just noticed the class imbalance in the data. There are far many control samples than PPROM or sPTD samples. If someone knows a good downsampling or upsampling strategies, balancing these numbers may be beneficial depending on the classifier. 

For a first pass, naive classifier, I will not employ these strategies.

Also, for sanity check, there should be 285+55 = **340 total training samples for Control vs. sPTD** and 285+95 = **380 total training samples for Control vs. PPROM** classifiers, respectively.

# Training Data (X, y) for Classifier 1: Control vs. sPTD

I will first extract data for the Control. vs sPTD classifier. 

```python
control_sPTD_annotation = train_annotation[~(train_annotation["Group"] == "PPROM")] # Select all rows not in PPROM group
assert control_sPTD_annotation.shape[0] == 340
control_sPTD_annotation.head()
```

```python
control_sPTD_SampleIDs = control_sPTD_annotation["SampleID"]
control_sPTD_data = train_data.loc[list(control_sPTD_SampleIDs)]
print("control_sPTD_data has {} samples and {} genes.".format(control_sPTD_data.shape[0], control_sPTD_data.shape[1]))
```

## Saving X Data for Classifier 1 

Similar to the test data, I will save the training X data as `X_train.csv` in the `data/processed/clf_1` folder. I will also save a `torch.Tensor` for easy access in `data/torch/clf_1` folder.

```python
control_sPTD_data.to_csv("../data/processed/clf_1/X_train.csv")
```

```python
control_sPTD_np = np.float32(control_sPTD_data.values)
X_control_sPTD = torch.from_numpy(control_sPTD_np)
torch.save(X_control_sPTD, "../data/torch/clf_1/X_train.pt")
```

## Saving y (label) Data for Classifier 1

I will save the entire control_sPTD_annotation dataframe as `y_train.csv` in `data/processed/clf_1` folder. For `torch.Tensor`, I will first encode the labels as 0 or 1 using `sklearn.preprocessing.LabelEncoder`, then save.

**Note**: The order of the labels is already the same as the rows of the training data! I am not mixing up labels for the samples. :)

Reference to `sklearn.preprocessing.LabelEncoder`: https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html

```python
control_sPTD_annotation.to_csv("../data/processed/clf_1/y_train.csv")
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(control_sPTD_annotation["Group"])
print(le.classes_)
le.transform(["Control", "sPTD"])
```

As you can see above, the `Control` has been encoded to class 0 and `sPTD` has been encoded to class 1. 

```python
control_sPTD_y_np = le.transform(control_sPTD_annotation["Group"])
assert control_sPTD_y_np.shape[0] == 340
```

```python
y_control_sPTD = torch.from_numpy(control_sPTD_y_np)
torch.save(y_control_sPTD, "../data/torch/clf_1/y_train.pt")
```
